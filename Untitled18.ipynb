```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x_ = df.drop('Purchased', axis=1)
y_ = df['Purchased']
```

```python
x_.head()
y_.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_, y_, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
x_train_new = scaler.transform(x_train)
x_test_new = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train
```

```python
x_train_new
```

```python
x_train_new = pd.DataFrame(x_train_new, columns=x_train.columns)
x_test_new = pd.DataFrame(x_test_new, columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logic = LogisticRegression()
logic_new = LogisticRegression()
```

```python
logic.fit(x_train, y_train)
logic_new.fit(x_train_new, y_train)
```

```python
y_pred = logic.predict(x_test)
y_pred_new = logic_new.predict(x_test_new)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("actual", accuracy_score(y_test, y_pred))
print("new_scale", accuracy_score(y_test, y_pred_new))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv = cross_val_score(logic, x_train, y_train, cv=8)
cv.mean()
```

```python
cv1 = cross_val_score(logic_new, x_train_new, y_train, cv=8)
cv1.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
df1 = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
df1.fit(x_train_new, y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred_new = df1.predict(x_test_new)
```

```python
print("actual", accuracy_score(y_test, y_pred))
print("new_scale", accuracy_score(y_test, y_pred_new))
```

```python
cv2 = cross_val_score(dt, x_train, y_train, cv=8)
print(cv2.mean())
cv3 = cross_val_score(df1, x_train_new, y_train, cv=8)
print(cv3.mean())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
